```python
%reload_ext autoreload
%autoreload 2

import numpy as np

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import colormaps as cms
mpl.rc_file(f"../matplotlibrc")
```

## phot

```python
l = np.array([30., 30., 29., 28., 28., 27., 27., 27., 27., 26., 26., 26.]) + np.log10(1/3)
m = np.logspace(2, 13, 12)

m_run = np.logspace(1.5, 12, 22)

p = np.polyfit(np.log10(m), l, 1)
l_run = np.polyval(p, np.log10(m_run))
```

```python
plt.plot(m, l, "-s", label='original step size used for fit (adjusted)')
plt.plot(m_run, l_run, '-+', label='production step size')
plt.xscale('log')
plt.axvline(10**1.5, color='k', ls='--')
plt.axvline(1e12, color='k', ls='--')
plt.legend();
```

```python
'[' + ', '.join([f'{lt:.3f}' for lt in l_run]) + ']'
```

```python
# test for certain mass
np.polyval(p, np.log10(1e7))
```

## elec

```python
l = np.array([27., 26., 27., 28., 28., 27., 27., 26., 26., 26., 25., 25.]) + 1. + np.log10(4)
m = np.logspace(6.5, 12, 12)

m_run = m

p = np.polyfit(np.log10(m), l, 3)
l_run = np.polyval(p, np.log10(m))
```

```python
plt.plot(m, l, "-s", label='original step size used for fit (adjusted)')
plt.plot(m_run, l_run, '-+', label='production step size')
plt.xscale('log')
plt.axvline(10**6.5, color='k', ls='--')
plt.axvline(1e12, color='k', ls='--')
plt.legend();
```

```python
# test for certain mass
np.polyval(p, np.log10(1e7))
```

```python
m_mid = np.sqrt(m[1:] * m[:-1])
l_mid_run = np.polyval(p, np.log10(m_mid))
```

```python
plt.plot(m, l, "-s", label='original step size used for fit (adjusted)')
plt.plot(m_run, l_run, '-+', label='production step size')
plt.plot(m_mid, l_mid_run, '+', label='production step size')
plt.xscale('log')
plt.axvline(10**6.5, color='k', ls='--')
plt.axvline(1e12, color='k', ls='--')
plt.legend();
```

```python
'[' + ', '.join([f'{lt:.3f}' for lt in l_mid_run]) + ']'
```
