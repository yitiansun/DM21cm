```python
%reload_ext autoreload
%autoreload 2

from PIL import Image
import numpy as np

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import colormaps as cms
mpl.rc_file('../matplotlibrc')
```

```python
def process_image(file_path, N):
    img = Image.open(file_path)
    img_gray = img.convert("L")
    img_resized = img_gray.resize((N, N))
    img_array = np.array(img_resized)
    img_normalized = 1 - img_array / 255

    return np.flipud(img_normalized)
```

```python
a = process_image('source.png', 128)
```

```python
plt.imshow(a, cmap=cms['binary'], origin='lower')
```

```python
plt.hist(a.flatten(), bins=30, range=(0, 1), density=True, log=True);
```

## Make reference array

```python
N = 128
a = process_image('source.png', N)
a = np.repeat([a], N, axis=0)
print(a.shape)
np.save(f'source_{N}.npy', a)
```
