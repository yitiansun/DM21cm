```python
def count_initial_identical_elements(arr):
    
    first_element = arr[0]
    count = 0
    
    for element in arr:
        if element == first_element:
            count += 1
        else:
            break
    
    return count

# Example usage
print(count_initial_identical_elements(z[:, 0]))  # Output: 3
```
